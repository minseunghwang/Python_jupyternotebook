```python
## 이거

import numpy as np
import pandas as pd

result_df = pd.DataFrame(columns=["영화제목", "각 영화별 평균평점"])
# 사용자 id부터 뽑아내 보아요!
df = pd.read_csv("./data/movielens/ratings.csv")
df2 = pd.read_csv("./data/movielens/movies.csv")

df3=pd.merge(df2, df, on="movieId", how="right")
movie_ids = df3['movieId'].unique()

for movie_id in movie_ids:
    movie_mean=df3.loc[df3["movieId"] == movie_id ,"rating"].mean()
    movie_title=df3.loc[df3["movieId"] == movie_id ,"title"].min()
    result_df.loc[movie_id] = [movie_title, movie_mean]
display(result_df)
```

```python
## 3 번 스나버젼
'''
3. 각 영화별 평균 평점을 출력하세요.
'''
import numpy as np
import pandas as pd

df1 = pd.read_csv('./data/movielens/movies.csv')
df2 = pd.read_csv('./data/movielens/ratings.csv')

df3 = pd.merge(df1, df2, on='movieId', how='right')
#display(df3)

result_df = pd.DataFrame(columns=['영화제목','영화평점'])
#display(result_df)


movie_idx = df3['movieId'].unique()
print("=====" *2)


for val in movie_idx:
    movie_mean = df3.loc[ df3['movieId']==val ,'rating'].mean()
    movie_title = df3.loc[ df3['movieId']==val ,'title'].max()
#    result_df.loc[val]= [movie_mean,df3[val,'title']]
    result_df.loc[val]= [movie_title, movie_mean]
    


display(result_df)
```
